```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('dataset_eg_acts.csv', sep=',',header=None, index_col =0)

data.hist(bins=list(np.linspace(0,5,500)))
plt.ylabel('Frequency')
plt.xlabel('Activation')
plt.title('Dataset Example Activations for the 50 units')
plt.show()



plt.show()
```
