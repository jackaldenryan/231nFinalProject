```python
import pickle

with open("./featurevis-loss-histories.pickle", "rb") as f:
    histories = pickle.load(f)
```

```python
import matplotlib.pyplot as plt

plt.title("Feature visualization of InceptionV1 units 0-50")

plt.xlabel("Step")
plt.ylabel("Average unit activation")

for history in histories:
    plt.plot(-history)

plt.show();
```
